# 1次元圧縮性Euler方程式
## 支配方程式
$\displaystyle \frac{\partial Q}{\partial t} + \frac{\partial F}{\partial x} = 0$

$\displaystyle e= \frac{1}{2}\rho u^2 + \frac{p}{\gamma-1}$

where  
$Q = \left[
\begin{matrix}
\rho \\ 
\rho u \\
e
\end{matrix}
\right]$

$F = \left[
\begin{matrix}
\rho u \\ 
\rho u^2 + p \\
(E + p)u
\end{matrix}
\right]$

$\rho$ is density of fluid, 
$u$ is flow velocity, 
$e$ is total energy per unit mass, and
$p$ is pressure.

## Flux jacobian
$\displaystyle \frac{\partial F}{\partial Q} = A = P\Lambda P^{-1}$

where 

$\displaystyle P=\left[
\begin{matrix}
1 & 1 & 1 \\ 
u-a & u & u+a \\
H-au & \frac{u^2}{2} & H+au
\end{matrix}
\right]$

$\displaystyle \Lambda = \left[
\begin{matrix}
u-a & 0 & 0 \\ 
0 & u & 0 \\
0 & 0 & u+a
\end{matrix}
\right]$

$\displaystyle P^{-1}=\left[
\begin{matrix}
\frac{1}{2}\left( \alpha_1 + \frac{u}{a} \right) & -\frac{1}{2}\left( \alpha_2 u + \frac{1}{a} \right) & \frac{\alpha_2}{2} \\ 
1-\alpha_1 & \alpha_2u & -\alpha_2 \\
\frac{1}{2}\left( \alpha_1 - \frac{u}{a} \right) & -\frac{1}{2}\left( \alpha_2 u - \frac{1}{a} \right) & \frac{\alpha_2}{2} \\ 
\end{matrix}
\right]$

$\displaystyle \alpha_1 = 
\frac{\left( \gamma - 1 \right) u^2}
{2a^2}$

$\displaystyle \alpha_2 = 
\frac{\left( \gamma - 1 \right)}
{a^2}$

$H$ is total enthalpy, $T$ is temperature, and $a$ is speed of sound.

$\displaystyle H = \rho c_p T = \frac{\gamma}{\gamma - 1} \rho T = \frac{\gamma}{\gamma - 1}p$  
$\displaystyle T = \frac{p}{\rho}$  
$\displaystyle a = \sqrt{ \gamma \frac{p}{\rho} }$


$\gamma = 1.4$ is heat capacity ratio.

## descritization
$\displaystyle \int_\Omega \left( \frac{\partial Q_i}{\partial t} + \frac{\partial F_i}{\partial x} \right) d\Omega= 0$

$\displaystyle \frac{\partial}{\partial t} \int_\Omega Q_i d\Omega + \int_{ \partial \Omega } F_i \vec{n} d\left( \partial \Omega \right)= 0$

$\displaystyle \frac{\partial}{\partial t} \bar{Q} \delta \Omega = - \sum_{ j \in \partial \Omega  } F_{ij} \vec{n}_j \delta (\partial \Omega)_j$

where  
$\displaystyle F_{ij} = \frac{1}{2} \left( F_i + F_j \right) - \frac{1}{2} A^{roe} \left( Q_j - Q_i \right)$

$A^{roe} = P\Lambda P^{-1}$  
Note that $P$, $\Lambda$ and $P^{-1}$ are calculated from the variable of roe average.

### Roe average
$\rho_{roe} = \sqrt{\rho_i \rho_j}$  
$\displaystyle u_{roe} = \frac{\sqrt{\rho_i} u_i + \sqrt{\rho_j} u_j}{\sqrt{\rho_i} + \sqrt{\rho_j} }$  
$\displaystyle H_{roe} = \frac{\sqrt{\rho_i} H_i + \sqrt{\rho_j} H_j}{\sqrt{\rho_i} + \sqrt{\rho_j} }$  
$\displaystyle a_{roe} = \sqrt{ \left( \gamma - 1 \right) \left( H_{roe} - \frac{u_{roe}^2}{2} \right) }$  

where H is total enthalpy per unit volume as below.  
$\displaystyle H = \frac{\left( e + p\right)}{\rho} = \frac{1}{\rho} \left( \frac{1}{2}\rho u^2 + \frac{p}{\gamma-1} + p  \right) = \frac{1}{2}u^{2} + \frac{\gamma}{\gamma-1}\frac{p}{\rho}$

### Coupling
#### Equations of Motion (1D)
$\displaystyle \frac{dX_B}{dt} = a_B X_B + b_B$  
where  
$\displaystyle X_B=\left[
\begin{matrix}
x_B \\ 
u_B
\end{matrix}
\right]$  
$\displaystyle a_B=\left[
\begin{matrix}
0 & 1 \\ 
-\omega_n^2 & -2 \zeta \omega_n
\end{matrix}
\right]$  
$\displaystyle b_B=\left[
\begin{matrix}
0 \\ 
F
\end{matrix}
\right]$  
$\displaystyle F=\frac{f}{m} = \frac{S_B p_B}{m}$  
$\displaystyle \omega_n = \sqrt{\frac{k}{m}}$  
$\displaystyle \zeta = \frac{c}{2\sqrt{mk}}$  

$\displaystyle X_B^{n+1} = \left[ \frac{I}{\Delta t} - \frac{1}{2} a_B \right]^{-1} \left\{ \left[ \frac{I}{\Delta t} + \frac{1}{2} a_B \right] X_B^n + \frac{b^{n+1} + b^n}{2} \right\}$  
$\displaystyle \left[ \frac{I}{\Delta t} - \frac{1}{2} a_B \right]^{-1} = \frac{1}{\left( -\zeta \omega_n + \frac{1}{2} \omega_n^2 + \frac{1}{2} \right) + \frac{1}{4} \omega_n^2 \Delta t } \left[ \begin{matrix}
1 - \zeta \omega_n \Delta t & -1 + \frac{\Delta t}{2} \\ 
-1+\frac{1}{2} \omega_n^2 \Delta t & 1 
\end{matrix}
\right]$  


#### boundary condition on object surface
$\displaystyle \frac{1}{\Delta t} \left[ \left( \rho u \right)_{B-\frac{1}{4}}^{n+1} \Delta x_B^{n+1} - \left( \rho u \right)_{B-\frac{1}{4}}^{n} \Delta x_B^{n} \right]$
$\displaystyle + \left\{ \left( \rho u^2 + p \right)_{B}^{n+\frac{1}{2}} - \left( \rho u^2 + p \right)_{B-\frac{1}{2}}^{n+\frac{1}{2}} \right\}$
$\displaystyle - \left[ \left( \rho u \right)_{B}^{n+\frac{1}{2}}  \delta x_{B}^{n+1} - \left( \rho u \right)_{B-\frac{1}{2}}^{n+\frac{1}{2}} \delta x_{B-\frac{1}{2}}^{n+1} \right] = 0$  

$\displaystyle \left( \cdot \right)_{B-\frac{1}{2}} = \frac{1}{2} \left\{ \left( \cdot \right)_{B} + \left( \cdot \right)_{B-1} \right\}$  
$\displaystyle \left( \cdot \right)_{B-\frac{1}{4}} = \frac{1}{4} \left\{ 3 \left( \cdot \right)_{B} + \left( \cdot \right)_{B-1} \right\}$  
$\displaystyle \left( \cdot \right)_{B-\frac{1}{4}}^{n+\frac{1}{2}} = \frac{1}{8} \left\{ 3 \left( \cdot \right)_{B}^{n+1} + 3 \left( \cdot \right)_{B}^{n} + \left( \cdot \right)_{B-1}^{n+1} + \left( \cdot \right)_{B-1}^{n} \right\}$  

$\displaystyle p_B^{n+1} = \left\{ \frac{1}{4} \alpha_{B-1}^n u_B^{n+1} \left( 3\Delta x_{B}^{n+1} - \delta x_B^{n+1} \right)+ \frac{1}{4} \left( \alpha_{B-1}^n u_B^{\left(n+1\right)^2}  + 1 \right) \right\}^{-1}$
$\left\{ -\left[ \frac{1}{4} \left(\rho u \right)_{B-1}^{n+1} \Delta x_B^{n+1} - \left(\rho u \right)_{B-\frac{1}{4}}^{n} \Delta x_B^{n} \right]
-\left[ \frac{1}{4} \left(\rho u^2 + p \right)_{B}^{n} - \frac{1}{2} \left( \rho u^2 + p \right)_{B-1}^{n+\frac{1}{2}}\right]
+\left[ \left(\rho u \right)_{B}^{n} \left( \frac{1}{2} \delta x_B^{n+1} - \frac{1}{4} \delta x_{B-\frac{1}{2}}^{n+1} \right) - \frac{1}{2} \left(\rho u \right)_{B-1}^{n+\frac{1}{2}} \delta x_{B-\frac{1}{2}}^{n+\frac{1}{2}} \right] \right\}$  

$\displaystyle \rho_B^{n+1} = \alpha_{B-1}^{n} p_B^{n+1} = \frac{p_B^{n+1}}{p_{B-1}^{n}} \rho_{B-1}^{n} $

#### SLAU2 algorithm
$\displaystyle \bar{\rho} = \frac{\rho_L + \rho_R}{2}$  
$\displaystyle \bar{c} = \frac{c_L + c_R}{2}$  
$\displaystyle \hat{M} = min\left( 1, \frac{1}{\bar{c}} \sqrt{ \frac{u_L^2 + u_R^2}{2}}\right)$  

$\displaystyle M_L = \frac{V_{nL}}{\bar{c}} = \frac{\vec{u_L} \bullet \vec{n_L}}{\bar{c}}$  
$\displaystyle M_R = \frac{V_{nR}}{\bar{c}} = \frac{\vec{u_R} \bullet \vec{n_R}}{\bar{c}}$  
$\displaystyle \left| \bar{V_n} \right| = \frac{\rho_L \left| V_{nL} \right| + \rho_R \left| V_{nR} \right| }{\rho_L + \rho_R}$  

$\displaystyle P_L^+ = ...$  
$\displaystyle P_R^- = ...$  
$\displaystyle P^{\pm}|_{\alpha} = \left\{ \begin{array}{ll} \frac{1}{2} \left( 1 \pm sign\left( M \right) \right) & \left( \left| M \right| \ge 1 \right) \\ \frac{1}{4} \left( M \pm 1 \right)^2 \left( 2 \mp M \right) \pm \alpha M \left( M^2 - 1\right)^2 & \left( \left|M\right| \lt 1 \right) \end{array} \right.$  
$\displaystyle \tilde{p} = \frac{p_L + p_R}{2} + \frac{P^+|_{\alpha=0} - P^-|_{\alpha=0}}{2} \left( p_{L} - p_{R} \right) + \gamma_{HR} \sqrt{\frac{u_L^2+u_R^2}{2}} \left( P^+|_{\alpha=0} + P^-|_{\alpha=0} - 1 \right) \bar{\rho} \bar{c}$  

$\displaystyle g = - \left[ max\left\{ min \left( M_L, 0 \right), -1 \right\} \bullet min\left\{ max\left( M_R, 0 \right), 1 \right\} \right]$  
$\displaystyle \chi = \left( 1 - \hat{M}^2 \right) $  

$\displaystyle \left| \bar{V_n} \right|^+ = (1-g)\left| \bar{V_n} \right| + g\left| V_{nL} \right|$  
$\displaystyle \left| \bar{V_n} \right|^- = (1-g)\left| \bar{V_n} \right| + g\left| V_{nR} \right|$  
$\displaystyle \dot{m} = \frac{1}{2} \left\{ \rho_L \left( V_{nL} + \left| \bar{V_n} \right|^+ \right) + \rho_R \left( V_{nR} + \left| \bar{V_n} \right|^- \right)\right\} - \frac{\chi}{\bar{c}} \left( p_R - p_L \right)$  
$\displaystyle E=\frac{\dot{m} + \left| \dot{m} \right|}{2}\psi^ + \frac{\dot{m} - \left| \dot{m} \right|}{2}\psi^- +\tilde{p}\vec{N} $  

```python
def slau2(qi, qj):
    rho_L, u_L, e_L, p_L = q2b(qi)
    rho_R, u_R, e_R, p_R = q2b(qj)
    n_L = 1
    n_R = 1
    H_L = ( g / g1 ) * p_L / rho_L + u_L**2 / 2
    H_R = ( g / g1 ) * p_R / rho_R + u_R**2 / 2

    VnL = u_L * n_L
    VnR = u_R * n_R

    c_L = sqrt(g * p_L / rho_L)
    c_R = sqrt(g * p_R / rho_R)
    
    bar_rho = (rho_L + rho_R) / 2
    # bar_c = (c_L + c_R) / 2 # SLAU
    
    # c12 is interfacial speed of sound  # SLAU2
    c_asterisk2_L = 2*g1/(g+1) * H_L
    c_asterisk2_R = 2*g1/(g+1) * H_R
    tilde_c_L = c_asterisk2_L / max(sqrt(c_asterisk2_L), abs(VnL))
    tilde_c_R = c_asterisk2_R / max(sqrt(c_asterisk2_R), abs(VnR))
    bar_c = min(tilde_c_L, tilde_c_R)
    
    hat_M = min(1, 1/bar_c * sqrt((u_L**2 + u_R**2)/2))
    
    M_L = VnL / bar_c
    M_R = VnR / bar_c
    
    alpha = 0
    if ( abs(M_L) > 1 ):
        Pplus_L = 1/2 * (1 + np.sign(M_L) )
    else:
        Pplus_L = 1/4 * (M_L + 1)**2 * (2 - M_L) + alpha * M_L * (M_L**2 - 1)**2
    
    if ( abs(M_R) > 1 ):
        Pminus_R = 1/2 * (1 - np.sign(M_R) )
    else:
        Pminus_R = 1/4 * (M_R - 1)**2 * (2 + M_R) - alpha * M_R * (M_R**2 - 1)**2
    
    # この値の計算はi-1, j+1が必要なため割愛
    # gamma_w = 1 - np.sign(min(phi_1, phi_2))
    gamma_w = 1    
    gamma_HR = max(0.2 , gamma_w)
    
    tilde_p = (p_L + p_R)/2 + (Pplus_L - Pminus_R)/2 * (p_L - p_R) + gamma_HR * sqrt( (u_L**2 + u_R**2 )/2 ) * (Pplus_L + Pminus_R - 1) * bar_rho * bar_c
    # tilde_p = (p_L + p_R)/2 + (Pplus_L - Pminus_R)/2 * (p_L - p_R) + gamma_HR * sqrt( (u_L**2 + u_R**2 )/2 ) * (Pplus_L + Pminus_R - 1) * bar_rho * c12
    g_slau = -( max( min(M_L, 0), -1 ) * min( max(M_R, 0), 1) )
    chi = (1 - hat_M)**2
    
    abs_Vn = ( rho_L * abs(VnL) + rho_R * abs(VnR) ) / ( rho_L + rho_R )
    Vn_plus = (1-g_slau)*abs_Vn + g_slau * abs(VnL)
    Vn_minus = (1-g_slau)*abs_Vn + g_slau * abs(VnR)
    
    m_dot = 1/2 * ( rho_L * (VnL + Vn_plus) + rho_R * (VnR - Vn_minus) - chi/bar_c * (p_R - p_L) )
    
    psi_L = np.array([1, u_L, H_L])
    psi_R = np.array([1, u_R, H_R])
    N = np.array([0, 1, 0])
    
    F = (m_dot + abs(m_dot))/2 * psi_L + (m_dot - abs(m_dot))/2 * psi_R + tilde_p * N

    return F  
```

```python
def fvs(qi, qj):   
    R, R_inv, L, L_abs = A_pm(qi)
    Ap = np.dot(np.dot(R, L+L_abs), R_inv)
    
    R, R_inv, L, L_abs = A_pm(qj)
    Am = np.dot(np.dot(R, L-L_abs), R_inv)
    
    F_plus = 1/2 * ( np.dot(Ap, qi) + np.dot(Am, qj))
    return F_plus

def A_pm(q):
    rho, u, e, p = q2b(q)
    H = (e + p) / rho
    c = sqrt(g1 * (H - u**2 / 2))
    b2 = g1 / c**2
    b1 = (b2 * u**2) / 2
    
    R = np.array([[1, 1, 1], [u-c, u, u+c], [H-u*c, u**2 / 2, H+u*c]])
    R_inv = np.array([[(b1+u/c)/2, (-b2*u-1/c)/2, b2/2],
                      [1-b1, b2*u, -b2], 
                      [(b1-u/c)/2, (-b2*u+1/c) / 2, b2/2]])
    
    L = np.array([[u-c, 0, 0],[0, u, 0],[0, 0, u+c]])
    L_abs = np.array([[abs(u-c), 0, 0],[0, abs(u), 0],[0, 0, abs(u+c)]])
    return R, R_inv, L, L_abs
    
```

```python
def get_flux(q):
    rho, u, e, p = q2b(q)
    f1 = rho * u
    f2 = rho * u**2 + p
    f3 = (e + p) * u
    f = np.array([f1, f2, f3])
    return f

def flux(Q, qB, qBLR):
    F = np.zeros((n+1, 3))
    if use_muscl:
        QL, QR = muscl(Q, qB, qBLR)
    
    for j in range(n+1):
        # print(j)
        i = j
        if ( j == obj_center_j ):
            F[j, :] = get_flux(qBLR[0])
            continue
            
        if ( j == 0 ):
            qi = qB[:, 0]
        else:            
            if use_muscl:
                qi = QL[i] # L
            else:
                qi = Q[j-1, :]

        if ( j == n ):
            qj = qB[:, 1]
        else:
            if use_muscl:
                qj = QR[i] # R
            else:
                qj = Q[j  , :]

        # print(f"edge:{j}")
        F[j, :] = slau2(qi, qj)
        # F[j, :] = fvs(qi, qj)
        # if ( qi[0] != qj[0] ):
            # print(j, qi, qj, F[j, :])
        # F[j, :] = roe_fds(qi, qj)
    # sqrt(-1)

    return F
```

```python
def bc_refrect(q):
    rho, u, e, p = q2b(q)
    qB = np.zeros(3)
    qB[0] = rho
    qB[1] = - rho * u
    qB[2] = e
    return qB
```

```python
def __bc_nonreflect(q1, q0, dx, dt):
    rho, u, e, p = q2b(q1)
    _, u0, _, p0 = q2b(q0)
    c = sqrt(g * p / rho)
    dpdx = (p - p0)/dx
    dudx = (u - u0)/dx
    
    lam1 = u-c
    lam2 = lam3 = lam4 = u
    lam5 = u+c
    
    L1 = lam1*(dpdx - rho * c * dudx)
    L2 = lam2*(c**2 * dpdx - dpdx)
    L5 = lam5*(dpdx + rho * c * dudx)
    
    d1 = 1/c**2 * (L2 + 1/2 * (L5 + L1) )
    d2 = 1/2 * (L5 + L1)
    d3 = 1/(2*rho*c) * (L5 - L1)
    
    qB = np.zeros(3)
    qB[0] = rho - dt * d1
    qB[1] = rho * u - dt * (u*d1 + rho*d3)
    qB[2] = e - dt * (1/2*u**2*d1 + d2/g1 + rho*u*d3)
    
    return qB

def __bc_inlet(q1):
    _, u, _, p = q2b(q1)
    rho_in, u_in, p_in = get_inlet_value()
    qB = np.zeros(3)
    if ( abs(u) >= 1 ): # supersonic flow    
        pass
    else:
        p_in = p # value from inside
        
    qB[0] = rho_in
    qB[1] = rho_in * u_in
    qB[2] = 1/2*rho_in * u_in**2 + p_in/g1
    return qB

def bc_io(n, q1=None, q0=None, dx=None, dt=None, p_inf=non_P):
    rho, u, e, p = q2b(q1)
    # else: # outlet
    qB = __bc_nonreflect(q1, q0, dx, dt)
    # check
    qB_p = g1 * (qB[2] - 1/2*qB[1]**2/qB[0])
    if ( qB[0] < 0 ) or ( qB[2] < 0 ) or (qB_p < 0):
        print("farfield")
        if ( abs(u) > 1 ):
            pass
        else:
            p = non_P
            e = 1/2 * rho * u**2 + p/g1
        qB = np.array([rho, rho*u, e])
    # print(qB)
    return qB

def bc_inflow(q1, dt, limit_t=0.5):
    rho, u, e, p = q2b(q1)
    rho_in, u_in, p_in = get_inlet_value()
    qB = np.zeros(3)
    if ( abs(u_in) >= 1): # supersonic
        pass
    else:
        p_in = p
    
    # if ( dt < limit_t):
        # u_in = u_in * (dt/limit_t)
        
    qB[0] = rho_in
    qB[1] = rho_in * u_in
    qB[2] = 1/2*rho_in * u_in**2 + p_in/g1
    return qB
```

```python
def calc_CFL(Q, dx, dt):
    CFL = 0
    for i in range(n):
        qi = Q[i, :]
        ( rho, u, e, p ) = q2b(qi)
        a = sqrt(g * p / rho)
        CFL_i = max(abs(u), abs(u-a), abs(u+a)) * dt / dx[i]
        CFL = max(CFL, CFL_i)
    return CFL

def set_dtau(Q, dx, dt):
    CFL = calc_CFL(Q, dx, dt)
    if ( CFL < 1 ):
        alpha = 2.1
    else:
        alpha = 1.22/CFL - 0.07*CFL + 0.95
    dtau = alpha*dt
    return dtau
```

```python
def get_roe_average(qi, qj):
    ( rho_i, u_i, e_i, p_i ) = q2b(qi)
    ( rho_j, u_j, e_j, p_j ) = q2b(qj)
    H_i = ( g / g1 ) * p_i / rho_i + u_i**2 / 2
    H_j = ( g / g1 ) * p_j / rho_j + u_j**2 / 2
    # print(qi, qj)
    rho = sqrt(rho_i * rho_j)
    u = ( sqrt(rho_i) * u_i + sqrt(rho_j) * u_j ) / (sqrt(rho_i) + sqrt(rho_j) )
    H = ( sqrt(rho_i) * H_i + sqrt(rho_j) * H_j ) / (sqrt(rho_i) + sqrt(rho_j) )
    a_sq = g1 * ( H - u**2 /2 )
    if a_sq < 0:
        print(qi, qj)
        print(rho, u, H, a_sq)
        raise ValueError
    # a = sqrt( g1 * ( H - u**2/2 ) )
    a = sqrt(a_sq)
    return rho, u, H, a
```

```python
plot_dt = 1
Nmax = 9 #int(0.5/dt)#int(Tmax / dt)
Pmax = int(Nmax / plot_dt )
fg = False
use_muscl = True#False#True#False
# paramter for rrk2
b1 = 17
b2 = -16
c2 = 1/32

def q_add_uB(q, uB):
    rho, u, e, p = q2b(q)
    return np.array([rho, rho*uB, e])

def calc_L(Qn1, Qn, dxn1, dxn, delta_x, dt, qBLR0, qBLR1):
    qBn  = np.zeros((3, 2))
    qBn1 = np.zeros((3, 2))
    # set boundary 
    qBn[:, 0]  = bc_refrect(Qn[0, :])
    qBn[:, 1]  = bc_refrect(Qn[n-1, :])
    qBn1[:, 0] = bc_refrect(Qn1[0, :])
    qBn1[:, 1] = bc_refrect(Qn1[n-1, :])
    
    Fn  = flux(Qn , qBn, qBLR0)
    Fn1 = flux(Qn1, qBn1, qBLR1)

    Qijn  = roeQ(Qn,  qBn, qBLR0)
    Qijn1 = roeQ(Qn1, qBn1, qBLR1)
    
    # Fn  = flux(Qn , qBn, qBw0)
    # Fn1 = flux(Qn1, qBn1, qBw1)
    # Qijn  = roeQ(Qn,  qBn, qBw0)
    # Qijn1 = roeQ(Qn1, qBn1, qBw1)
    
    Fn12 = (Fn + Fn1) / 2
    Qij12 = (Qijn + Qijn1) / 2

    L = np.zeros((n, 3))
    for i in range(n):
        if ( i == obj_i_L ) or ( i == obj_i_R ):
            continue
        
        j = i
        m1 = j
        p1 = j + 1
        L[i, :] = 1/dt * ( Qn1[i, :] - dxn[i]/dxn1[i] * Qn[i, :] ) \
                + 1/dxn1[i] * ( Fn12[p1, :] - Fn12[m1, :] ) \
                - 1/dt * ( delta_x[p1]/dxn1[i] * Qij12[p1, :] - delta_x[m1]/dxn1[i] * Qij12[m1, :])

    return L

def rrk2(Q):
    def dot(a, b):
        return np.sum(a*b)
        
    Q_store = np.zeros((n, 3, Pmax))
    x_store = np.zeros((n, Pmax))
    x_e_0 = None
    retry = 50
    fixed_grid=fg
    epsilon = 10**(-14)
    _, x_e_n, dxn, _, u_B0 = update_x(Q, Q, x_e_0=None, u_B=None)
    u_B1 = u_B0
    Qn = Q # 初期値設定1
    qBLR0 = np.array([Qn[obj_i_L], Qn[obj_i_R]])
    for k in range(Nmax):        
        # dtau = set_dtau(Qn, dxn1, dt) # 内部反復の時間刻み幅の決定
        dtau = set_dtau(Qn, dxn, dt) # 内部反復の時間刻み幅の決定
        # print(np.max(delta_x))
        # iteration
        for count in range(retry + 1):
            if ( count == 0 ):
                Qn1 = Qn # 次段階の初期値はQn
                qBLR1 = qBLR0 #np.array([Qn1[obj_i-1], Qn1[obj_i+1]])
            
            # 運動方程式による物体位置/格子の更新
            x_c_n1, x_e_n1, dxn1, delta_x, u_B1_new = update_x(Qn, Qn1, x_e_n, u_B0, qBLR0, qBLR1)
            du_B1 = u_B1_new - u_B1
            u_B1 = u_B1_new

            # print(k, calc_CFL(Qn1, dxn1, dt))
            # print(u_B1)
            # rrk2 による流れ場の物理量の更新
            s1 = -dtau * calc_L(Qn1, Qn, dxn1, dxn, delta_x, dt, qBLR0, qBLR1) # s1 == g1
            s2 = -dtau * calc_L(Qn1 + c2 * s1, Qn, dxn1, dxn, delta_x, dt, qBLR0, qBLR1) # s1 == g1
            s3 = b1*s1 + b2*s2
            ss11 = dot(s1, s1)
            ss13 = dot(s1, s3)
            ss33 = dot(s3, s3)
            if ( ss33 != 0 ):
                dQ = (2*s1*ss13 - s3*ss11) / ss33
            else:
                dQ = 0
            Qn1 = Qn1 + dQ
            # 物体表面における物理量(境界条件)の更新
            qBLR1 = update_surface_q(Qn1, Qn, u_B1, dxn1, dxn, delta_x, dt, qBLR0)
            Qn1[obj_i_L] = qBLR1[0]
            Qn1[obj_i_R] = qBLR1[1]
            # 残差の評価
            res_u = np.sqrt(du_B1**2)
            res_avg = np.sqrt(np.sum((dQ)**2) / (3*n))
            res_max = np.sqrt(np.max(dQ**2))
            res = max(res_avg, res_max, res_u)
            
            if ( res < epsilon ):
                break
                
            if ( count == retry ):
                print(k, "not converge")
                print(res)
                # return Q_store, x_store

        if ( k % plot_dt == 0 ):
            print(k, count, res, calc_CFL(Qn1, dxn1, dt), np.max(np.abs(Qn1[:, 1] / Qn1[:, 0]) ))
            m = int(k / plot_dt)
            Q_store[:, 0, m] = Qn1[:, 0]
            Q_store[:, 1, m] = Qn1[:, 1]
            Q_store[:, 2, m] = Qn1[:, 2]
            x_store[:, m] = x_c_n1

        # prepare for next time step
        Qn = Qn1
        x_e_n = x_e_n1
        dxn = dxn1
        u_B0 = u_B1
        qBLR0 = qBLR1
        # print(x_c_n1[obj_i_R])
        # print(qBLR1[0])
        # print(Qn[obj_left_j-1:obj_right_j+1])
        # print(qBLR1[1])
        # sqrt(-1)
        if ( qBLR1[1][0] < 0 ):
            print(u_B1)
            print(Qn1)
            print(qBLR1)
            sqrt(-1)

        
        # print(delta_x[obj_left_j-1:obj_right_j+2] / dt)
            
    return Q_store, x_store

Qsr, xsr = rrk2(Q)
    
```
