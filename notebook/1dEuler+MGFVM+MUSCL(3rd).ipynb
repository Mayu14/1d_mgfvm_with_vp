# 1次元圧縮性Euler方程式
## 支配方程式
$\displaystyle \frac{\partial Q}{\partial t} + \frac{\partial F}{\partial x} = 0$

$\displaystyle e= \frac{1}{2}\rho u^2 + \frac{p}{\gamma-1}$

where  
$Q = \left[
\begin{matrix}
\rho \\ 
\rho u \\
e
\end{matrix}
\right]$

$F = \left[
\begin{matrix}
\rho u \\ 
\rho u^2 + p \\
(E + p)u
\end{matrix}
\right]$

$\rho$ is density of fluid, 
$u$ is flow velocity, 
$e$ is total energy per unit mass, and
$p$ is pressure.

## Flux jacobian
$\displaystyle \frac{\partial F}{\partial Q} = A = P\Lambda P^{-1}$

where 

$\displaystyle P=\left[
\begin{matrix}
1 & 1 & 1 \\ 
u-a & u & u+a \\
H-au & \frac{u^2}{2} & H+au
\end{matrix}
\right]$

$\displaystyle \Lambda = \left[
\begin{matrix}
u-a & 0 & 0 \\ 
0 & u & 0 \\
0 & 0 & u+a
\end{matrix}
\right]$

$\displaystyle P^{-1}=\left[
\begin{matrix}
\frac{1}{2}\left( \alpha_1 + \frac{u}{a} \right) & -\frac{1}{2}\left( \alpha_2 u + \frac{1}{a} \right) & \frac{\alpha_2}{2} \\ 
1-\alpha_1 & \alpha_2u & -\alpha_2 \\
\frac{1}{2}\left( \alpha_1 - \frac{u}{a} \right) & -\frac{1}{2}\left( \alpha_2 u - \frac{1}{a} \right) & \frac{\alpha_2}{2} \\ 
\end{matrix}
\right]$

$\displaystyle \alpha_1 = 
\frac{\left( \gamma - 1 \right) u^2}
{2a^2}$

$\displaystyle \alpha_2 = 
\frac{\left( \gamma - 1 \right)}
{a^2}$

$H$ is total enthalpy, $T$ is temperature, and $a$ is speed of sound.

$\displaystyle H = \rho c_p T = \frac{\gamma}{\gamma - 1} \rho T = \frac{\gamma}{\gamma - 1}p$  
$\displaystyle T = \frac{p}{\rho}$  
$\displaystyle a = \sqrt{ \gamma \frac{p}{\rho} }$


$\gamma = 1.4$ is heat capacity ratio.

## descritization
$\displaystyle \int_\Omega \left( \frac{\partial Q_i}{\partial t} + \frac{\partial F_i}{\partial x} \right) d\Omega=\int_\Omega \tilde{ \nabla } \tilde{ F }_i d\Omega=0$

where
$\displaystyle \tilde{ \nabla } = \left( \frac{\partial }{\partial t}, \frac{\partial }{\partial x}\right)$
$\displaystyle \tilde{ F }_i = \left( Q_i, F_i \right)$

$\displaystyle \int_\Omega \tilde{ \nabla } \tilde{ F }_i d\Omega = \int_{\partial \Omega} \tilde{ F }_i \vec{n} d\left( \partial \Omega \right) = 0$  
where 
$\vec{ n } =\left( n_t, n_x \right)$

$\displaystyle \int_{\partial \Omega} \tilde{ F }_i \vec{n} d\left( \partial \Omega \right)
= \sum_{ j \in \partial \Omega  } \tilde{F}_{ij} \vec{n}_j \delta (\partial \Omega)_j
= \sum_{ j \in \partial \Omega  } \tilde{F}_{ij} \vec{\tilde{n}}_j = 0
$  
where $\vec{\tilde{n}}_j = \vec{n}_j \delta (\partial \Omega)_j$

$\displaystyle j=1 : \left( t=n, x=i \right)$  
$\displaystyle \tilde{F}_{i1} \vec{\tilde{n}}_1 = \left( Q_{i}^{n}, F_{i}^{n} \right) \cdot \left( -\Delta x_i^{n}, 0 \right) = -Q_{i}^{n}\Delta x_i^{n}$  

$\displaystyle j=2 : \left( t=n+\frac{1}{2}, x=i+\frac{1}{2} \right)$  
$\displaystyle \tilde{F}_{i2} \vec{\tilde{n}}_2$
$= \left( Q_{i+\frac{1}{2}}^{n+\frac{1}{2}}, F_{i+\frac{1}{2}}^{n+\frac{1}{2}} \right) \cdot \left( -\delta x_{i+\frac{1}{2}}^{n+\frac{1}{2}}, \Delta t \right) = - Q_{i+\frac{1}{2}}^{n+\frac{1}{2}} \delta x_{i+\frac{1}{2}}^{n+\frac{1}{2}} + F_{i+\frac{1}{2}}^{n+\frac{1}{2}} \Delta t$  

$\displaystyle j=3 : \left( t=n, x=i \right)$  
$\displaystyle \tilde{F}_{i3} \vec{\tilde{n}}_3 = \left( Q_{i}^{n+1}, F_{i}^{n+1} \right) \cdot \left( \Delta x_i^{n+1}, 0 \right) = Q_{i}^{n+1}\Delta x_i^{n+1}$  


$\displaystyle j=4 : \left( t=n+\frac{1}{2}, x=i-\frac{1}{2} \right)$  
$\displaystyle \tilde{F}_{i4} \vec{\tilde{n}}_4$
$= \left( Q_{i-\frac{1}{2}}^{n+\frac{1}{2}}, F_{i-\frac{1}{2}}^{n+\frac{1}{2}} \right) \cdot \left( \delta x_{i-\frac{1}{2}}^{n+\frac{1}{2}}, -\Delta t \right) = Q_{i-\frac{1}{2}}^{n+\frac{1}{2}} \delta x_{i-\frac{1}{2}}^{n+\frac{1}{2}} - F_{i-\frac{1}{2}}^{n+\frac{1}{2}} \Delta t$  

where  
$\displaystyle Q^{n+\frac{1}{2}} = \frac{1}{2}\left( Q^n + Q^{n+1} \right)$  
$\displaystyle F^{n+\frac{1}{2}} = \frac{1}{2}\left( F^n + F^{n+1} \right)$  

$\displaystyle Q_{\frac{i+j}{2}} = Roe\left( Q_i , Q_j \right) = \left[
\begin{matrix}
\rho_{roe} \\ 
\rho_{roe} u_{roe} \\
e_{roe}
\end{matrix}
\right]$ 

$\displaystyle F_{\frac{i+j}{2}} = \frac{1}{2} \left( F_i + F_{j} \right) - \frac{1}{2} A_{roe} \left( Q_j - Q_i \right)$  

$e_{roe}=\frac{1}{\gamma}H_{roe} + \frac{1}{2}\rho_{roe}u_{roe}^2$

$A^{roe} = P\Lambda P^{-1}$  
Note that $P$, $\Lambda$ and $P^{-1}$ are calculated from the variable of roe average.

### Roe average
$\rho_{roe} = \sqrt{\rho_i \rho_j}$  
$\displaystyle u_{roe} = \frac{\sqrt{\rho_i} u_i + \sqrt{\rho_j} u_j}{\sqrt{\rho_i} + \sqrt{\rho_j} }$  
$\displaystyle H_{roe} = \frac{\sqrt{\rho_i} H_i + \sqrt{\rho_j} H_j}{\sqrt{\rho_i} + \sqrt{\rho_j} }$  
$\displaystyle a_{roe} = \sqrt{ \left( \gamma - 1 \right) \left( H_{roe} - \frac{u_{roe}^2}{2} \right) }$  

## Time step
$\displaystyle L\left( Q_i^{n + 1}, Q_i^{n}\right) = 
\frac{1}{\Delta t} \left( 
        Q_i^{n+1}
        - \frac{ \Delta x_i^{n} }{ \Delta x_i^{n+1} } Q_i^{n}
    \right)
    + 
    \frac{1}{\Delta x_i^{n+1}} \left(
        F_{i + \frac{1}{2}}^{ n + \frac{1}{2}}
      - F_{i - \frac{1}{2}}^{ n + \frac{1}{2}}
    \right)
    - \frac{1}{\Delta t} \left(
        \frac{ \delta x_{i + \frac{1}{2}}^{ n + \frac{1}{2}} }
             {\Delta x_{i}^{ n + 1}}
        Q_{i + \frac{1}{2}}^{ n + \frac{1}{2}}
      - \frac{ \delta x_{i - \frac{1}{2}}^{ n + \frac{1}{2}} }
             {\Delta x_{i}^{ n + 1}}
        Q_{i - \frac{1}{2}}^{ n + \frac{1}{2}}
    \right)
$

$\displaystyle \frac{\partial Q}{\partial \tau} = -L\left( Q_i^{n+1}, Q_i^{n}\right)$

### explicit euler
$\displaystyle Q_i^{\left( n+1 \right)\left( \nu+1 \right)} = Q_i^{\left( n+1 \right)\left( \nu \right)} -\Delta \tau L\left( Q_i^{\left( n+1 \right)\left( \nu \right)}, Q_i^{n}\right)$  
where
$Q_i^{\left( n+1 \right)\left( 0 \right)} = Q_i^{n}$

where $\left( , \right)$ means dot product of all region.  
$\left( g_a, g_b\right) = \sum_{i=1}^{N} \sum_{j=1}^{3} g_{aij} g_{bij}$  
Note that the result is scalar variable.

where  
$b_1 = 17$  
$b_2 = -16$  
$c_2 = \frac{1}{32}$  
$\alpha = \left\{
\begin{array}{ll}
2.1 & (CFL \leq 1) \\
\frac{1.22}{CFL}-0.07CFL+0.95 & \left( 1.0 < CFL < 10.0 \right)
\end{array}
\right.
$  

### RRK2
$\displaystyle g_1 = -\Delta \tau L\left( Q_i^{\left( n+1 \right)\left( \nu \right)}, Q_i^{n}\right)$  
$\displaystyle g_2 = -\Delta \tau L\left( Q_i^{\left( n+1 \right)\left( \nu \right)} + c_2g_1, Q_i^{n}\right)$  
$g_3=b_1g_1 + b_2g_2$  
$Q_i^{\left( n+1 \right)\left( \nu + 1\right)} = Q_i^{\left( n+1 \right)\left( \nu \right)} + \frac{2g_1\left( g_1, g_3\right) - g_3\left( g_1, g_1\right)}{\left( g_3, g_3\right)}$  
$Q_i^{\left( n+1 \right)\left( 0 \right)} = Q_i^{n}$

## MUSCL (Monotonical Upwind Scheme for COnservation Law)
### TVD Condition (Total Variation Diminishing)
以下で定義される全変動について  
$TV\left(Q^n\right) = \sum \left| Q_{i+1}^n - Q_{i}^n\right|$  
次の関係を成立させることが数値振動を抑えるための条件  
$TV\left(Q^{n+1}\right) \le TV\left(Q^{n}\right)$  


### 界面における物理量の再構築
$\displaystyle u_{i+\frac{1}{2}}^L = u_{i} + \frac{1-k}{4} \phi\left( \frac{1}{r_{i}} \right) \left( u_{i} - u_{i-1} \right)  + \frac{1+k}{4} \phi\left( r_{i} \right) \left( u_{i+1} - u_i \right) $  
$\displaystyle u_{i-\frac{1}{2}}^L = u_{i} - \frac{1-k}{4} \phi\left( \frac{1}{r_{i}} \right) \left( u_{i} - u_{i-1} \right)  - \frac{1+k}{4} \phi\left( r_{i} \right) \left( u_{i+1} - u_i \right) $  
$k=\frac{1}{3}$のとき3次精度となる．
これらを基礎変数$\left( \rho, u, p \right)$に対し適用する．  
$\displaystyle r_{i} = \frac{u_i - u_{i-1}}{u_{i+1} - u_{i}}$  
また，$\phi$はTVD条件を満足させるためのLimiter．  

#### minmod limiter
$minmod\left(r_i, 1\right) = sgn\left(r_i\right) max\left[0, \left\{ min\left( \left| r_i \right|, sgn\left( r_i \right) \times 1 \right)\right\} \right]$  
where
$minmod\left(\frac{p}{q}, 1\right)q = minmod\left(p, q\right)$

hense
$\displaystyle u_{i+\frac{1}{2}}^L = u_{i} + \frac{1-k}{4} minmod\left( \Delta_i^-, b\Delta_i^+ \right) + \frac{1+k}{4} minmod \left( \Delta_i^+, b\Delta_i^- \right) $  
$\displaystyle u_{i-\frac{1}{2}}^R = u_{i} - \frac{1-k}{4} minmod\left( \Delta_i^+, b\Delta_i^- \right) + \frac{1+k}{4} minmod \left( \Delta_i^-, b\Delta_i^+ \right) $  
where  
$ \Delta_i^- = u_i - u_{i-1}, \Delta_i^+ = u_{i+1} - u_{i}$  

#### characteristic variables
基礎変数に対して適用すると数値振動が発生したため，特性変数に対して処理してみる

\right]

```python

def q2b(q):
    rho = q[0]
    u = q[1] / q[0]
    e = q[2] 
    p = g1 * ( e - (rho * u**2 / 2) )
    return rho, u, e, p

def q2f(q):
    (rho, u, e, p) = q2b(q)
    F = np.zeros(3)
    F[0] = rho * u
    F[1] = rho * u**2 + p
    F[2] = (e + p) * u
    return F

def entropy_fix(a, epsilon=0.1):
    abs_a = abs(a)
    if ( abs_a < epsilon ):
        mod_a = 1/2* (abs_a**2/epsilon + epsilon)
    else:
        mod_a = abs_a
    return mod_a

def get_roe_average(qi, qj):
    ( rho_i, u_i, e_i, p_i ) = q2b(qi)
    ( rho_j, u_j, e_j, p_j ) = q2b(qj)
    H_i = ( g / g1 ) * p_i / rho_i + u_i**2 / 2
    H_j = ( g / g1 ) * p_j / rho_j + u_j**2 / 2
    # print(qi, qj)
    rho = sqrt(rho_i * rho_j)
    u = ( sqrt(rho_i) * u_i + sqrt(rho_j) * u_j ) / (sqrt(rho_i) + sqrt(rho_j) )
    H = ( sqrt(rho_i) * H_i + sqrt(rho_j) * H_j ) / (sqrt(rho_i) + sqrt(rho_j) )
    a_sq = g1 * ( H - u**2 /2 )
    if a_sq < 0:
        print(qi, qj)
        print(rho, u, H, a_sq)
        raise ValueError
    # a = sqrt( g1 * ( H - u**2/2 ) )
    a = sqrt(a_sq)
    return rho, u, H, a

def roe_fds(qi, qj):
    rho, u, H, a = get_roe_average(qi, qj)
    
    P = np.zeros((3,3))
    invP = np.zeros((3,3))
    L = np.zeros((3,3))
    
    a1 = ( g1 * u**2 ) / (2 * a**2)
    a2 = g1 / a**2
    
    P[0, 0] = 1
    P[0, 1] = 1
    P[0, 2] = 1
    P[1, 0] = u - a
    P[1, 1] = u
    P[1, 2] = u + a
    P[2, 0] = H - a * u
    P[2, 1] = u**2 / 2
    P[2, 2] = H + a * u

    L[0, 0] = entropy_fix(u - a)
    L[1, 1] = entropy_fix(u)
    L[2, 2] = entropy_fix(u + a) 
    
    invP[0, 0] = 1/2 * (a1 + u/a)
    invP[0, 1] = -1/2 * (a2*u + 1/a)
    invP[0, 2] = a2 / 2
    invP[1, 0] = 1 - a1
    invP[1, 1] = a2 * u
    invP[1, 2] = -a2
    invP[2, 0] = 1/2*(a1 - u/a)
    invP[2, 1] = -1/2 * (a2*u - 1/a)
    invP[2, 2] = a2 / 2

    dq = qj - qi
    dF = np.dot(P, np.dot(L, np.dot(invP, dq)))
    
    F_i = q2f(qi)
    F_j = q2f(qj)
    #print(qi, qj)
    #print(F_i, F_j)
    roeF = 1/2 * (F_i + F_j) - 1/2*dF
    return roeF  
```

```python
def Q2B(Q):
    Rho = Q[:, 0]
    U = Q[:, 1] / Q[:, 0]
    E = Q[:, 2] 
    P = g1 * ( E - (Rho * U**2 / 2) )
    return Rho, U, E, P
    

def muscl_org(Q, qB):
    def minmod(p, q, b):
        # r = p / q
        # return np.sign(r) * max(0, min(abs(r), np.sign(r)))
        # return np.sign(p) * max(0, min(abs(p), np.sign(p)*q*b))
        # return np.sign(p) * max(0, min(abs(p), np.sign(p)*q))
        return 0
        
    def minmod_LR(x_m1, x_0, x_p1, k=1/3):
        # b = (3-k) / (1-k)
        b = 1
        delta_m = x_0 - x_m1
        delta_p = x_p1 - x_0
        x_L = x_0 + (1-k)/4 * minmod(delta_m, delta_p, b) + (1+k)/4 * minmod(delta_p, delta_m, b)
        x_R = x_0 - (1-k)/4 * minmod(delta_p, delta_m, b) - (1+k)/4 * minmod(delta_m, delta_p, b)
        return x_L, x_R
    
    Rho, U, E, P = Q2B(Q)
    # Rho, M, E = Q[:, 0], Q[:, 1], Q[:, 2]
    nQ = np.zeros((n, 3, 2))
    
    for i in range(n):
        if ( i == 0 ):
            rho_m1, u_m1, _, p_m1 = q2b(qB[:, 0])
            # rho_m1, m_m1, e_m1 = qB[0, 0], qB[1, 0], qB[2, 0]
        else:
            rho_m1 = Rho[i-1]
            u_m1 = U[i-1]
            p_m1 = P[i-1]
            # m_m1 = M[i-1]
            
        rho_0 = Rho[i]
        u_0 = U[i]
        p_0 = P[i]
        
        if ( i == n-1 ):
            rho_p1, u_p1, _, p_p1 = q2b(qB[:, 1])
        else:
            rho_p1 = Rho[i+1]
            u_p1 = U[i+1]
            p_p1 = P[i+1]
        
        rho_L, rho_R = minmod_LR(rho_m1, rho_0, rho_p1)
        u_L, u_R = minmod_LR(u_m1, u_0, u_p1)
        p_L, p_R = minmod_LR(p_m1, p_0, p_p1)
        nQ[i, 0, 1] = rho_L
        nQ[i, 0, 0] = rho_R
        nQ[i, 1, 1] = rho_L*u_L
        nQ[i, 1, 0] = rho_R*u_R
        nQ[i, 2, 1] = 1/2*rho_L*u_L**2 + p_L/g1
        nQ[i, 2, 0] = 1/2*rho_R*u_R**2 + p_R/g1
    return nQ

def flux(Q, qB):
    nQ = muscl(Q, qB)
    F = np.zeros((n+1, 3))
    for j in range(n+1):
        i = j
        if ( j == 0 ):
            qi = qB[:, 0]
        else:
            qi = nQ[i-1, :, 1] # L
        
        if ( j == n ):
            qj = qB[:, 1]
        else:
            qj = nQ[i, :, 0] # R
        
        F[j, :] = roe_fds(qi, qj)

    return F

def roeQ(Q, qB):
    nQ = muscl(Q, qB)
    roeQ = np.zeros((n+1, 3))
    for j in range(n+1):
        i = j
        if ( j == 0 ):
            qi = qB[:, 0]
        else:
            qi = nQ[i-1, :, 1]
        
        if ( j == n ):
            qj = qB[:, 1]
        else:
            qj = nQ[i, :, 0]
        
        rho, u, H, a = get_roe_average(qi, qj)
        m = rho*u
        p = a**2 * rho / g
        e = rho*H - p
        # e = 1/2*rho*u**2 + rho/g1 * (H - u**2/2) # H = (e+p)/rho
        roeQ[j, 0] = rho
        roeQ[j, 1] = m
        roeQ[j, 2] = e
        # print(rho, u, m, H, a, e)
    return roeQ
```

```python
def muscl(Q, qB):
    def minmod(p, q, b):
        # r = p / q
        # return np.sign(r) * max(0, min(abs(r), np.sign(r)))
        return np.sign(p) * max(0, min(abs(p), np.sign(p)*q*b))
        # return np.sign(p) * max(0, min(abs(p), np.sign(p)*q))
        # return 0
        
    def minmod_LR(x_m1, x_0, x_p1, k=1/3):
        b = (3-k) / (1-k)
        # b = 1
        delta_m = x_0 - x_m1
        delta_p = x_p1 - x_0
        x_L = x_0 + (1-k)/4 * minmod(delta_m, delta_p, b) + (1+k)/4 * minmod(delta_p, delta_m, b)
        x_R = x_0 - (1-k)/4 * minmod(delta_p, delta_m, b) - (1+k)/4 * minmod(delta_m, delta_p, b)
        return x_L, x_R
    
    # Rho, U, E, P = Q2B(Q)
    Rho, M, E = Q[:, 0], Q[:, 1], Q[:, 2]
    nQ = np.zeros((n, 3, 2))
    
    for i in range(n):
        if ( i == 0 ):
            # rho_m1, u_m1, _, p_m1 = q2b(qB[:, 0])
            rho_m1, u_m1, p_m1 = qB[0, 0], qB[1, 0], qB[2, 0]
        else:
            rho_m1 = Rho[i-1]
            u_m1 = M[i-1]#U[i-1]
            p_m1 = E[i-1]#P[i-1]
            
        rho_0 = Rho[i]
        u_0 = M[i]#U[i]
        p_0 = E[i]#P[i]
        
        if ( i == n-1 ):
            # rho_p1, u_p1, _, p_p1 = q2b(qB[:, 1])
            rho_p1, u_p1, p_p1 = qB[0, 1], qB[1, 1], qB[2, 1]
        else:
            rho_p1 = Rho[i+1]
            u_p1 = M[i+1]#U[i+1]
            p_p1 = E[i+1]#P[i+1]
        
        rho_L, rho_R = minmod_LR(rho_m1, rho_0, rho_p1)
        u_L, u_R = minmod_LR(u_m1, u_0, u_p1)
        p_L, p_R = minmod_LR(p_m1, p_0, p_p1)
        nQ[i, 0, 1] = rho_L
        nQ[i, 0, 0] = rho_R
        nQ[i, 1, 1] = u_L#rho_L*u_L
        nQ[i, 1, 0] = u_R#rho_R*u_R
        nQ[i, 2, 1] = p_L#1/2*rho_L*u_L**2 + p_L/g1
        nQ[i, 2, 0] = p_R#1/2*rho_R*u_R**2 + p_R/g1
    return nQ
```

```python
def update_dx_from_edge(x_e):
    dx = np.diff(x_e)
    return dx

def update_x_from_edge(x_e):
    x_c = np.zeros(n)
    for i in range(n):
        x_c[i] = (x_e[i] + x_e[i+1])/2
    return x_c
   
def update_x(t, x_e_0=None, omega=1, coef=0.4, fixed=False):
    x_e = np.zeros(n+1)
    avg_dx = non_L / n
    avg_rad = omega*t
    avg_rad_n = omega*(t-1)
    
    delta_x = np.zeros(n+1)
    for j in range(n+1):
        if fixed or ( j == 0 ) or ( j == n ):
            x_e[j] = j * avg_dx
        else:
            x_e[j] = j * avg_dx + coef*( avg_dx )*np.sin( avg_rad   + j )
            x_e_n  = j * avg_dx + coef*( avg_dx )*np.sin( avg_rad_n + j)
            delta_x[j] = x_e[j] - x_e_n
        
    x_c = update_x_from_edge(x_e)
    dx = update_dx_from_edge(x_e)
    if ( t == 0 ):
        delta_x = np.zeros(n+1)
    return x_c, x_e, dx, delta_x
```

```python
def calc_CFL(Q, dx, dt):
    CFL = 0
    for i in range(n):
        qi = Q[i, :]
        ( rho, u, e, p ) = q2b(qi)
        a = sqrt(g * p / rho)
        CFL_i = max(abs(u), abs(u-a), abs(u+a)) * dt / dx[i]
        CFL = max(CFL, CFL_i)
    return CFL

def set_dtau(Q, dx, dt):
    CFL = calc_CFL(Q, dx, dt)
    if ( CFL < 1 ):
        alpha = 2.1
    else:
        alpha = 1.22/CFL - 0.07*CFL + 0.95
    dtau = alpha*dt
    return dtau
```

```python
def bc_refrect(q):
    rho, u, e, p = q2b(q)
    qB = np.zeros(3)
    qB[0] = rho
    qB[1] = - rho * u
    qB[2] = e
    return qB
```

## Time step
$\displaystyle L\left( Q_i^{n + 1}, Q_i^{n}\right) = 
\frac{1}{\Delta t} \left( 
        Q_i^{n+1}
        - \frac{ \Delta x_i^{n} }{ \Delta x_i^{n+1} } Q_i^{n}
    \right)
    + 
    \frac{1}{\Delta x_i^{n+1}} \left(
        F_{i + \frac{1}{2}}^{ n + \frac{1}{2}}
      - F_{i - \frac{1}{2}}^{ n + \frac{1}{2}}
    \right)
    - \frac{1}{\Delta t} \left(
        \frac{ \delta x_{i + \frac{1}{2}}^{ n + \frac{1}{2}} }
             {\Delta x_{i}^{ n + 1}}
        Q_{i + \frac{1}{2}}^{ n + \frac{1}{2}}
      - \frac{ \delta x_{i - \frac{1}{2}}^{ n + \frac{1}{2}} }
             {\Delta x_{i}^{ n + 1}}
        Q_{i - \frac{1}{2}}^{ n + \frac{1}{2}}
    \right)
$

```python
plot_dt = 4
Nmax = 200#int(0.5/dt)#int(Tmax / dt)
Pmax = int(Nmax / plot_dt )
fg = False

def calc_L(Qn1, Qn, dxn1, dxn, delta_x, dt):
    qBn  = np.zeros((3, 2))
    qBn1 = np.zeros((3, 2))
    # set boundary 
    qBn[:, 0]  = bc_refrect(Qn[0, :])
    qBn[:, 1]  = bc_refrect(Qn[n-1, :])
    qBn1[:, 0] = bc_refrect(Qn1[0, :])
    qBn1[:, 1] = bc_refrect(Qn1[n-1, :])
    
    Fn  = flux(Qn , qBn)
    Fn1 = flux(Qn1, qBn1)
    Fn12 = (Fn + Fn1) / 2
    
    Qijn  = roeQ(Qn,  qBn)
    Qijn1 = roeQ(Qn1, qBn1)
    Qij12 = (Qijn + Qijn1) / 2
    L = np.zeros((n, 3))
    for i in range(n):
        m1 = i
        p1 = i + 1
        L[i, :] = 1/dt * ( Qn1[i, :] - dxn[i]/dxn1[i] * Qn[i, :] ) \
                + 1/dxn1[i] * ( Fn12[p1, :] - Fn12[m1, :] ) \
                - 1/dt * ( delta_x[p1]/dxn1[i] * Qij12[p1, :] - delta_x[m1]/dxn1[i] * Qij12[m1, :])
    return L

def rrk2(Q):
    def dot(a, b):
        return np.sum(a*b)
        
    Q_store = np.zeros((n, 3, Pmax))
    x_e_0 = None
    retry = 100
    fixed_grid=fg
    epsilon = 10**(-14)
    _, x_e_n, dxn, _ = update_x(t=0*dt, x_e_0=None, fixed=fixed_grid)
    Qn = Q
    for k in range(Nmax):
        _, x_e_n1, dxn1, _ = update_x(t=(k+1)*dt, x_e_0=x_e_0, fixed=fixed_grid)
        delta_x = x_e_n1 - x_e_n
        dtau = set_dtau(Qn, dxn1, dt)
        # print(np.max(delta_x))
        # iteration
        for count in range(retry + 1):
            if ( count == 0 ):
                Qn1 = Qn
                
            s1 = -dtau * calc_L(Qn1, Qn, dxn1, dxn, delta_x, dt) # s1 == g1
            s2 = -dtau * calc_L(Qn1 + c2 * s1, Qn, dxn1, dxn, delta_x, dt) # s1 == g1
            s3 = b1*s1 + b2*s2
            ss11 = dot(s1, s1)
            ss13 = dot(s1, s3)
            ss33 = dot(s3, s3)
            dQ = (2*s1*ss13 - s3*ss11) / ss33
            Qn1 = Qn1 + dQ
            
            res_avg = np.sqrt(np.sum((dQ)**2) / (3*n))
            res_max = np.sqrt(np.max(dQ**2))
            res = max(res_avg, res_max)

            if ( res < epsilon ):
                break
                
            if ( count == retry ):
                print(k, "not converge")
                print(res)
                return Q_store

        if ( k % plot_dt == 0 ):
            print(k, count, res)
            m = int(k / plot_dt)
            Q_store[:, 0, m] = Qn1[:, 0]
            Q_store[:, 1, m] = Qn1[:, 1]
            Q_store[:, 2, m] = Qn1[:, 2]

        # prepare for next time step
        Qn = Qn1
        x_e_n = x_e_n1
        dxn = dxn1
            
    return Q_store
    
Qsr = rrk2(Q)
    
```
